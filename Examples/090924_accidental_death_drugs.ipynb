```python
import pandas as pd
# import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('../../../mimiLabs_Yubin/Accidental_Drug_Related_Deaths_2012-2022.csv')
data.head()
```

```python
data.dropna(subset=['Age', 'Sex', 'Heroin'], inplace=True)
data.isnull().any()
```

```python
# data.head()
data_columns = data.columns
data_columns
```

```python
data['Date'] = pd.to_datetime(data['Date']).dt.year
data.head()
```

```python
data['Date'].value_counts().sort_index().plot(kind='line')
```

```python
data['Sex'].value_counts().plot(kind='bar')
```

```python
data_by_gender = data.groupby('Sex').count()
data_by_gender
```

```python
data_gender_columns = [ 'Heroin', 'Cocaine', 'Fentanyl', 'Oxycodone', 'Oxymorphone', 'Hydrocodone', 'Benzodiazepine', 'Methadone', 'Amphet', 'Tramad', 'Other', 'Any Opioid']
data_by_gender = data_by_gender[data_gender_columns]
data_by_gender.plot(kind='bar', figsize=(15, 6))
```

```python
# df_columns = {'Heroin': ['count'], 'Cocaine': ['count']}
columns_to_check = [ 'Heroin', 'Cocaine', 'Fentanyl', 'Oxycodone', 'Oxymorphone', 'Hydrocodone', 'Benzodiazepine', 'Methadone', 'Amphet', 'Tramad', 'Other', 'Any Opioid']

df_data = data[columns_to_check]
df_data.head()
```

```python
res = df_data.aggregate(['count'])
res.plot(kind='bar', figsize=(15, 6), xlabel='Drugs', ylabel='Count', title='Number of deaths by drug')
```
